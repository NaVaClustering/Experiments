```python
import re
def remove_emails(dataframe):
    no_emails = re.sub("\S*@\S*\s?","",dataframe)
    no_emails = re.sub("From:","",no_emails)
    no_emails = re.sub("Subject:","",no_emails)
    return no_emails
```
